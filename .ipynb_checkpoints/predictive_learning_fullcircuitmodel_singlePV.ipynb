```python
%autosave 1
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import scipy.stats as stats
```

## PV with square activation function and square in learning rule

```python
def phi_square(_x):
    x = np.array([_x])
    x[x<=0] = 0
    x[x>0] = x[x>0]**2 
    x[x>20] = 20
    return x[0]
```

```python
def phi_RL(_x):
    """rectified linear activation function"""
    x = np.array([_x])
    x[x<=0] = 0
    x[x>20] = 20
    return x[0]
```

```python
def phi(x_,case=1):
    x = np.array([x_])
    if case == 1:
        x[x<=0] = 0
        x[x>20] = 20
    return x[0]
    
```

```python
def run_network(mean= 5.0, sigma= 0.5, wP= 6.0, w_init_PV = 0.01, w_init_R = 0.1, plastic=True, plastic_PV=True, patho=False, rectified=1):
    """
    Parameters
    ----------
    mean : float
        mean of the stimulus distribution
    sigma: float
        standard deviation of the stimulus distribution
    wP : float
        connection weight from R to PV and from SST to PV
    w_init_PV : float
        initial weight from Y1 to PV
    w_init_R : float
        initial weight from Y1 to R
    plastic : boolean
        truth value indicates whether plasticity is switched on or off   
        
    Returns
    ----------
    E_avg : float
        mean rate of error neuron
    PV_avg : float
        mean rate of PV
    SST_avg : float
        mean rate of SST
    R_avg : float
        mean rate of R
    wPX1_monitor : array of length t
        weight from X1 to PV over the course of the simulation
    wRX1_monitor : array of length t
        weight from X1 to R over the course of the simulation
    rE_monitor : array of length t
        rate of error neuron over the course of the simulation
    rS_monitor : array of length t
        rate of error neuron over the course of the simulation
    rP_monitor : array of length t
        rate of error neuron over the course of the simulation
    rY1 : array of length t
        whisker deflections over the course of the simulation
    """
    
    #sim parameters
    dt = 1
    T = 150000 # number of time steps

    # time constants
    tau_P = 1.0
    tau_S = 1.0
    tau_E = 1.0
    tau_R = 1.0

    # weights
    # to Representation neuron
    wRX1 = np.array([w_init_R]) # small intitial weight
    #wRY1 = np.array([1.0])
    wRE_P = np.array([1.0])
    wRE_N = np.array([1.0])
    wPX1 = np.array([w_init_PV]) # small intitial weights

    
    # weights in positive PE circuit
    # to PV
    wPY1 = np.array([wP]) # small intitial weights
    wPS_P = np.array([-wP])
    
    # to SST
    wSR_P = np.array([1.0])

    # to E
    wEP = np.array([1.0])
    wES_P = np.array([-1.0])
    wEY1_P = np.array([1.0])
    
    # weights in negative PE circuit
    # to PV
    wPR = np.array([wP])
    wPS_N = np.array([-wP])
    
    # to SST
    wSY1_N = np.array([1.0])

    # to E
    wEP = np.array([1.0])
    wES_N = np.array([-1.0])
    wER_N = np.array([1.0])

    # nudging parameter
    beta_R = 0.1 
    beta_P = 0.1


    # inputs
    # sound
    rX1 = np.ones((T))

    # whisker
    Y1_mean = mean
    Y1_sigma = sigma

    rY1 = Y1_sigma*np.random.randn(T)+Y1_mean

    #learning rate
    eta_R = 0.1
    eta_P = 0.001

    #initial rates
    rP = 0.0#phi_square(rB[0])
    rS_P = 0.0
    rE_P = 0.0
    rE_patho = 0.0
    rS_N = 0.0
    rE_N = 0.0
    
    rR = 0.0

    # monitors
    rP_monitor = np.empty((T))
    rE_P_monitor = np.empty((T))
    rE_patho_monitor = np.empty((T))
    rS_P_monitor = np.empty((T))
    rE_N_monitor = np.empty((T))
    rS_N_monitor = np.empty((T))
    
    rR_monitor = np.empty((T))

    wPX1_monitor = np.empty((T))
    wRX1_monitor = np.empty((T))
    
    PV_in_target = np.empty((T))
    PV_in_wX = np.empty((T))
    PV_in = np.empty((T))
    PV_deltaw = np.empty((T))
    

    if patho == True: 
        healthy = 0.0
    else: 
        healthy = 1.0

    for t in range(T):

        
        drS_P = (-rS_P + phi_RL(wSR_P * rR))/tau_S
        drP = (-rP + phi_square(((1-beta_P)*(wPX1 * rX1[t]) + beta_P*(wPY1 * rY1[t] + wPS_P * rS_P + wPR * rR + wPS_N * rS_N))))/tau_P
        
        drE_P = (-rE_P + phi(1.0/(1.0 + wEP * rP) * (wEY1_P * rY1[t] + wES_P * rS_P),case=rectified))/tau_E
            
        drE_patho = (-rE_patho + 1.0/(1.0 + 2.0) * (wEY1_P * rY1[t] + wES_P * rS_P))/tau_E
        
        drS_N = (-rS_N + phi_RL(+ wSY1_N * rY1[t]))/tau_S
        drE_N = (-rE_N + phi(1.0/(1.0 + wEP * rP) * (wER_N * rR + wES_N * rS_N),case=rectified))/tau_E

        drR = (-rR + phi_RL((wRX1 * rX1[t]) + (wRE_P * rE_P) - (wRE_N * rE_N)))/tau_R # no convex combination, input from error neuron
        
        """drS_P = (-rS_P + phi_RL(wSR_P * rR))/tau_S
        drP_P = (-rP_P + phi_square(((1-beta_P)*(wPX1_P * rX1[t]) + beta_P*(wPY1_P * rY1[t] + wPS_P * rS_P))))/tau_P
        drE_P = (-rE_P + (1.0/(1.0 + wEP_P * rP_P) * (wEY1_P * rY1[t] + wES_P * rS_P)))/tau_E
        drE_patho = (-rE_patho + 1.0/(1.0 + 2.0) * (wEY1_P * rY1[t] + wES_P * rS_P))/tau_E
        
        drS_N = (-rS_N + phi_RL(+ wSY1_N * rY1[t]))/tau_S
        drP_N = (-rP_N + phi_square(((1-beta_P)*(wPX1_N * rX1[t]) + beta_P*(wPR_N * rR + wPS_N * rS_N))))/tau_P
        drE_N = (-rE_N + (1.0/(1.0 + wEP_N * rP_N) * (wER_N * rR + wES_N * rS_N)))/tau_E

        drR = (-rR + phi_RL((wRX1 * rX1[t]) + (wRE_P * rE_P)))/tau_R # no convex combination, input from error neuron
        #drR = (-rR + phi_RL((wRX1 * rX1[t]) + (wRE_P * rE_P) - (wRE_N * rE_N)))/tau_R # no convex combination, input from error neuron
        """
        #drR = (-rR + (wRX1 * rX1[t]) + patho*(wRE * rE_patho) + healthy*(wRE * rE))/tau_R # no convex combination, input from error neuron
        
        
        # store monitors
        rP_monitor[t] = rP
        rS_P_monitor[t] = rS_P
        rE_P_monitor[t] = rE_P
        rE_patho_monitor[t] = rE_patho
        rS_N_monitor[t] = rS_N
        rE_N_monitor[t] = rE_N
        rR_monitor[t] = rR
        wPX1_monitor[t] = wPX1
        wRX1_monitor[t] = wRX1
        PV_in_target[t] = beta_P*(wPY1 * rY1[t] + wPS_P * rS_P)
        PV_in_wX[t] = (1-beta_P)*(wPX1 * rX1[t])


        # rate changes
        rP += dt*drP
        rS_P += dt*drS_P
        rE_P += dt*drE_P
        rS_N += dt*drS_N
        rE_N += dt*drE_N
        rR += dt*drR
        rE_patho += dt*drE_patho

        # weight changes
        if t > 100:
            wPX1 += plastic_PV * eta_P * ((rP - ((wPX1*rX1[t])**2)) * rX1[t])
            #print(plastic * eta_P * ((rP_P - ((wPX1*rX1[t])**2)) * rX1[t]))
            wRX1 += plastic * eta_R * ((rR - (wRX1*rX1[t])) * rX1[t])
        PV_deltaw[t] = rP - ((wPX1*rX1[t])**2)
        PV_in[t] = (wPX1*rX1[t])**2
    
    results = {
    'PV_P_avg' : np.mean(rP_monitor[100:]),
    'SST_P_avg': np.mean(rS_P_monitor[100:]),
    'E_P_avg' : np.mean(rE_P_monitor[100:]),
    'R_avg' : np.mean(rR_monitor[100:]),
    'SST_N_avg' : np.mean(rS_N_monitor[100:]),
    'E_N_avg' : np.mean(rE_N_monitor[100:]),
    'wPX1' : wPX1_monitor, 
    'wRX1' : wRX1_monitor, 
    'rE_P' : rE_P_monitor, 
    'rS_P' : rS_P_monitor, 
    'rP_P' : rP_monitor, 
    'rE_N' : rE_N_monitor, 
    'rS_N' : rS_N_monitor, 
    'rR' : rR_monitor, 
    'rY1' : rY1,
    'PV_in_target' : PV_in_target,
    'PV_in_wX' : PV_in_wX,
    'PV_deltaw' : PV_deltaw,
    'PV_in' : PV_in
    }
    
    E_patho_avg = np.mean(rE_patho_monitor[100:])

    return results
    
```

```python
Y1_mean = 5.0
Y2_mean = 5.0
Y1_sigma = 1.0
Y2_sigma = 0.4
results1 = run_network(mean= Y1_mean,sigma=Y1_sigma, patho=False,rectified=0)
results2 = run_network(mean= Y2_mean,sigma=Y2_sigma, patho=False,rectified=0)

```
